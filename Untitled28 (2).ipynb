*Chi-Square test of independence between the 'department' and 'education' columns in the dataset to determine whether these two categorical variables are statistically related.*

```python
from scipy.stats import chi2_contingency
print(chi2_contingency(pd.crosstab(data['department'], data['education']))[1])
```

```python
import matplotlib.pyplot as plt

filtered_data = data[(data['no_of_trainings'] > 5) &
                     (data['previous_year_rating'] > 4) &
                     (data['awards_won'] >= 1) &
                     (data['is_promoted'] == 1) &
                     ((data['length_of_service'] > 12) & (data['avg_training_score'] >= 60))]

dept_counts = filtered_data['department'].value_counts()

plt.bar(dept_counts.index, dept_counts.values, color='blue')
plt.xlabel("Department")
plt.ylabel("Count")
plt.title("Employees Promoted per Department")
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

q = data[
    (data['no_of_trainings'] > 5) &
    (data['previous_year_rating'] > 4) &
    (data['awards_won'] >= 1) &
    (data['is_promoted'] == 1) &
    ((data['length_of_service'] > 12) & (data['avg_training_score'] >= 60))
]

sns.histplot(q['avg_training_score'], kde=True, bins=20)

plt.title("Distribution of Avg Training Score for Selected Employees")
plt.xlabel("Avg Training Score")
plt.ylabel("Density")
plt.show()
```
