```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"/content/Wholesale customers data.csv")
df
```

```python
df.drop(['Channel', 'Region'], axis=1, inplace=True)
```

```python
array=df.values
```

```python
array
```

```python
stscaler=StandardScaler().fit(array)
X=stscaler.transform(array)

X
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,6))
dendo=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customer Data')
plt.ylabel('Eucl Distance')
plt.show()
```

```python
len(set(dendo['color_list']))-1
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(2, 11):  # iterating over possible number of clusters from 2 to 10
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)  # assuming 'x' is your data
    wcss.append(kmeans.inertia_)

# You can then plot the WCSS to find the optimal number of clusters (elbow method)
import matplotlib.pyplot as plt
plt.plot(range(2, 11), wcss)
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Assuming X is your data
model = KMeans(n_clusters=3)  # Correct parameter name is 'n_clusters'
model.fit(X)
```

```python
import pandas as pd

# Assuming 'model' is already fitted and 'group' is the cluster labels (model.labels_)
group = model.labels_

# Create DataFrame with column name 'group'
group_num = pd.DataFrame(group, columns=['group'])

# Display the DataFrame
print(group_num)
```

```python
from sklearn.metrics import silhouette_score

# Assuming 'X' is your data and 'groups' are the cluster labels
groups = model.labels_  # Get cluster labels from the fitted model

# Calculate the silhouette score
score = silhouette_score(X, groups)

print("Silhouette Score:", score)
```
